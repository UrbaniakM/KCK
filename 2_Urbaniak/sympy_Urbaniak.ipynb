```python
%matplotlib inline
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
# Definicje zmiennych
x, y = sym.symbols("x y")

W = (x**3 + 3*x**2 - 10*x + 24)
```

```python
x_val = np.arange(-5,5.1,0.1)
#Wx = [W.evalf(x_val[i]) for i in range(len(x_val))]
Wx = [W.subs(x,i) for i in x_val]
W_plot = plt.subplot(111)
W_plot.plot(x_val,Wx)
W_plot.grid(color='#D3D3D3', linestyle='dotted', linewidth=1)
W_plot.set_xlim(-5,5)
W_plot.set_title('y = ' + str(W))
W_plot.set_xticks([x for x in range(-5, 6)])
print('Pierwiastki wielomianu: ')
sym.solve(sym.Eq(W),x)
```

```python
Equations = [sym.Eq(x**2 + 3*y,10), sym.Eq(4*x - y**2,-2)]
solutionsEq = sym.solve(Equations,[x,y])
print('Ilosc rozwiazan: ' + str(len(solutionsEq)))
print('Rozwiazania ukladu rownan: ')
solutionsEq
```

```python
print('Rozwiązania układu równań w postaci numerycznej: ')
for solution in solutionsEq:
    print('x = ' + str(solution[0].evalf()) + ', y = ' + str(solution[1].evalf()))
```
